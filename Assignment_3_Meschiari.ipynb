```python
plt.subplots(figsize=(12,8))
plt.plot(df_film["imdb_rating"], df_film["title"])
plt.title("Hollywood Blockbusters IMDB Ratings")
plt.ylabel('title')
plt.xlabel('rating')
plt.xlim(2,10)
plt.ylim(5,30)
plt.show()
```

```python
plt.subplots(figsize=(12,8))
plt.plot(df_film["studio"], df_film["worldwide_gross"])
plt.title("Highest Grossing Studios")
plt.ylabel('studio')
plt.xlabel('grossing')
plt.xlim(1,8)
plt.ylim(5,15)
plt.xticks([0,2,4,6,8,10])
plt.show()
```

```python
plt.subplots(figsize=(12,8))
plt.scatter(df_film["imdb_rating"],df_film["studio"])
plt.ylabel('studio')
plt.xlabel('imdb ratings')
plt.xlim(2,10)
plt.ylim(1,8)
plt.xticks([0,2,4,6,8,10])
plt.show()
```

```python
df_film.plot(x ='studio', y='imdb_rating', kind = 'bar', figsize = (14,8))
plt.xlabel("Hollywood Studios")
plt.ylabel("IMDB Ratings")
plt.title("Which Studio produces the best movies?")
plt.xlim(1,10)
plt.ylim(5,9)
plt.show()
```
